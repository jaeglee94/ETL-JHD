```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import sqlite3
```

```python
#connect
conn = sqlite3.connect("../movies.db")
```

```python
curs = conn.cursor()
curs.execute("select sql from sqlite_master where type='table'")
rows = curs.fetchall()

for row in rows:
    print(row)
```

```python
# URL of page to be scraped
url = 'https://www.amctheatres.com/movies'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html'
soup = BeautifulSoup(response.text, 'html')
print(type(soup))
```

```python
print(soup.prettify())
```

```python
# Retrieve the parent divs
results = soup.find_all('div', class_="PosterSlide")
print(results[25])
```

```python
for result in results:
    title = result.find('a')['aria-label']
    image = result.find('source')['srcset']
    theater = "AMC Theatres"
    rowInsert=(title,image,theater)
    curs.execute('insert into amc values (?,?,?)',rowInsert)
    conn.commit()
```

```python
curs.execute("select * from amc")
rows = curs.fetchall()

for row in rows:
    print(row)
```
